```python
from anomalib.data import Folder, MVTecAD
from anomalib.deploy import ExportType, OpenVINOInferencer
from anomalib.engine import Engine
from anomalib.models import Cflow, EfficientAd, Patchcore
from anomalib.utils.visualization import ImageResult
from torchvision.transforms import v2
import os
```

```python
augmentations = v2.Compose(
    [
        v2.RandomHorizontalFlip(
            p=0.5
        ),  # Randomly flip images horizontally with 50% probability
        v2.RandomVerticalFlip(
            p=0.2
        ),  # Randomly flip images vertically with 20% probability
        v2.RandomRotation(
            degrees=30
        ),  # Randomly rotate images within a range of ±30 degrees
        v2.RandomResizedCrop(
            size=(256, 256), scale=(0.8, 1.0)
        ),  # Randomly crop and resize images
        v2.ColorJitter(
            brightness=0.1, contrast=0.1, saturation=0.1, hue=0.1
        ),  # Randomly adjust colors
        v2.RandomGrayscale(
            p=0.1
        ),  # Convert images to grayscale with 10% probability
    ]
)
datamodule = Folder(
    name="custom_folder",
    root="/content/drive/MyDrive/processed",
    normal_dir="train/good",
    abnormal_dir="test/quexian",
    mask_dir="ground_truth/quexian",
    train_batch_size=1,
)
datamodule.setup()
model = EfficientAd(imagenet_dir="data/imagenette2")
# self.model = Cflow(
#     anomaly_source_path=os.path.join(
#         self.data_dir, "images", "processed", "test"
#     ),  # 异常样本路径
#     task="segmentation",
# )
# TODO
engine = Engine(max_epochs=300, accelerator="cuda")
engine.fit(datamodule=datamodule, model=model)
```

```python
import os
engine.export(
    model=model,
    export_type=ExportType.OPENVINO,
    export_root="/content/drive/MyDrive/model"
)
```
